```python
from gym_service import ServiceReward, ServiceEnv, FinishingCondition
from requests import Session
import numpy as np
```

```python
class ConstantReward(ServiceReward):

    def calculate_reward(self, prev_obs_space, curr_obs_space):
        return 0

class Finisher(FinishingCondition):

    def terminate_or_truncate(self, observation, reward):
        return False, False
```

```python
session = Session()
service_env = ServiceEnv(casca_url="http://localhost:8080/v0", session=session, reward=ConstantReward(), finisher=Finisher(), wait_seconds=0, slos=["meanFPS"], config_params=["EncodingThreadCount"])
```

```python
service_env.reset()
```

```python
service_env.step(np.array([1]))
```

```python
session.close()
```
